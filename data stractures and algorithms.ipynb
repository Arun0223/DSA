```python

def print_items(n):
    for i in range(n):
        print(i)
    print('*******************************')
    for j in range(n):
        print(j)
print_items(10)
```

```python
==> The above code running n times + n times
n+n = 2n
firts assumption we need to drop constant
2n ==> O(n)
```

```python
def sqaure(n):
    for i in range(n):
        for j in range(n):
            print(i,j)
```

```python
==> The above code will produce the output of n*n ==>n ^2 ==> O(n2)
for example we got answer like O(n^3) we will consider as O(n^2)
```

```python
==> Drop non-dominants
```

```python
def sqaure(n):
    for i in range(n):
        for j in range(n):
            print(i,j)
    for k in range(n):
        print(k)
```

```python
sqaure(10)
```

```python
==> Now it will become O(n^2) + O(n)
==> for example we have 1000 iterations
==> here O(n) won't effect calculation so O(n) is a non-dominant so we can drop simply
```
